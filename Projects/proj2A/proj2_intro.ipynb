## This Assignment 
Welcome to the second project of Data 100! By this point in the course you have completed many assignments that are focused on a specific part of the data science lifecycle. From exploratory data analysis with the bike sharing dataset to feature engineering and modeling in spam/ham classification, you've learned the basics of many different segments of the data science lifecycle. In this project we combine all the individual segments and present a holistic representation of the entire lifecycle through the perspective of the NYC Taxi Rides dataset. We hope that with this project you will learn what the lifecycle looks like from beginning to end.

In this project we will be using the NYC Taxi Rides dataset to predict ride duration given two external datasets: the NYC weather dataset and the NYC accidents dataset. This project will be broken down into multiple parts, each covering a core portion of the lifecycle. You have worked with many of these portions in previous assignments, but we will introduce a few more. The project will be broken down as follows:

### Part A (Due Monday, 11/26, 11:59 PM)
1. Data Wrangling
    - Querying for data from a database using SQL
    - Basic data cleaning using SQL/Pandas
2. EDA, Visualization, Feature Engineering
    - Exploratory data analysis of the NYC Taxi dataset
    - Generate visualizations to gain insight on the dataset

There is also an ethics notebook required for Part A.

### Part B (Due Friday, 11/30, 11:59 PM)
3. External Datasets
    - Exploratory data analysis of the NYC accidents dataset
    - Join the taxi dataset with NYC accidents dataset, make connection and observation
4. Feature Engineering and Model Fitting
    - Feature engineering based on gained insights
    - Modeling with sklearn data pipelines
    - Submission to Kaggle

There is also an extras notebook required for Part B.

## Structure of the Notebooks
You may have noticed that in previous assignments we only used one notebook. You may also have noticed that these notebooks can get really long, making debugging, reproducibility, legibility, and portability harder. While it may not have made as much sense to split up some of the homeworks given that we are only looking at one dataset at one part of the lifecycle, it is generally good practice to divide up your notebooks into clear, organized compartments. Especially with this project, which is quite long, we will make good habits by separating our work into parts. In order to use our manipulated datasets across different parts of the project, we will be saving our tables into `.hdf` files and loading them when needed. We will help you through this in each of the parts, but feel free to save your own tables and use them wherever you like.
